```python
figsize = (10, 10)
```

```python
bgr_image = cv2.imread("grass.jpg")
bgr_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=figsize)
plt.imshow(bgr_image)
plt.title("Original Image")
plt.show()
```

```python
rgb_green = np.uint8([[[0, 255, 0]]])
hsv_green = cv2.cvtColor(rgb_green, cv2.COLOR_RGB2HSV)[0, 0, :]
print(hsv_green)
```

```python
thresholding_schematics = cv2.cvtColor(
    cv2.imread("hsv_th.png"), cv2.COLOR_BGR2RGB
)
plt.figure(figsize=figsize)
plt.imshow(thresholding_schematics)
plt.title("Thresholding Schematics")
plt.show()
```

```python
hsv_image = cv2.cvtColor(bgr_image, cv2.COLOR_RGB2HSV)

lower_th = hsv_green - np.array([70, 200, 200])
upper_th = hsv_green + np.array([30, 0, 0])

mask = cv2.inRange(hsv_image, lower_th, upper_th)

plt.figure(figsize=figsize)
plt.imshow(mask)
plt.title("Resulted Mask")
plt.show()
```

```python
rgb_result = cv2.bitwise_and(bgr_image, bgr_image, mask=mask)

plt.figure(figsize=figsize)
plt.imshow(rgb_result)
plt.title("Output Image")
plt.show()
```
